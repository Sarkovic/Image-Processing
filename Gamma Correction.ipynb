# Gamma Correction

Gamma correction is a non-linear operation used to encode and decode luminance or tristimulus values in images or video. It adjusts the brightness and contrast of an image by modifying the pixel intensity values.

## Equation

Gamma correction is performed using the following equation:

I_out = I_in ^ (1 / ?)

Where:
- `I_in` is the input intensity value of a pixel.
- `I_out` is the output intensity value of a pixel.
- `?` (gamma) is the correction factor. 

## Usage

Gamma correction is commonly used in image processing to adjust the brightness and contrast of images. By changing the value of the gamma parameter, you can control the overall brightness and contrast of the image. A gamma value greater than 1 will increase the brightness and decrease the contrast, while a gamma value less than 1 will decrease the brightness and increase the contrast.

Gamma correction is particularly useful in situations where the brightness and contrast of images need to be adjusted to improve visibility or enhance certain features.

To learn more about gamma correction, visit the [Gamma Correction Website](https://en.wikipedia.org/wiki/Gamma_correction).

## Gamma Value 2

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

read_img_dir = r"D:\Research\Git\CV-ImageProcessing\image\inputCT.png"  
read_img = cv2.imread(read_img_dir)
g = cv2.cvtColor(read_img, cv2.COLOR_BGR2GRAY)  # GRAYSCALE format
gamma = np.array(255*(g / 255) ** 2.0, dtype='uint8')  # applying gamma correction using value = 2


plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.imshow(cv2.cvtColor(read_img, cv2.COLOR_BGR2RGB))
plt.title('Original Image')
plt.axis('off')


plt.subplot(1, 2, 2)
plt.imshow(gamma, cmap='gray')
plt.title('Gamma Corrected Image')
plt.axis('off')

plt.show()
```
