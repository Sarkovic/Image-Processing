# Circle binary mask

A circle binary mask is a binary image where pixels inside a specified circle are set to one (white), and pixels outside the circle are set to zero (black).

## Formula

The equation of a circle in Cartesian coordinates is given by:

```
(x - xc)^2 + (y - yc)^2 = r^2
```

Where:
- `(xc, yc)` are the coordinates of the circle's center.
- `r` is the radius of the circle.
- `(x, y)` are the coordinates of any point on the circle.

To create a circle binary mask, each pixel in the image is checked against this equation. If the pixel is within the circle (satisfies the equation), it is set to one (white); otherwise, it is set to zero (black).



To learn more about gamma correction, visit the [Circle Binary Mask Website](https://de.mathworks.com/help/images/ref/circles2mask.html).

````python
import cv2
import numpy as np
import matplotlib.pyplot as plt

path = r"D:\Research\Git\CV-ImageProcessing\image\inputCT.png"
image = cv2.imread(path, 0)

mask = np.zeros(image.shape, dtype=np.uint8)
mask = cv2.circle(mask, (175, 175), 177, (255, 255, 255), -1)

result = cv2.bitwise_and(image, mask)

plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.imshow(image, cmap='gray')
plt.title('Original Image')
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(result, cmap='gray')
plt.title('Masked Result')
plt.axis('off')

plt.show()
````
