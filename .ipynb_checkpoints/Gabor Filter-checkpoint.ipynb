# Gabor Filter

The Gabor filter is a linear filter used in image processing for texture analysis, edge detection, feature extraction, and more. It is particularly effective in capturing texture information at different scales and orientations.

## Equation

The Gabor filter is defined by the following equation:

```
G(x, y; f, θ, σ, γ, φ) = exp(-(x'^2 + γ^2 * y'^2) / (2 * σ^2)) * cos(2 * π * f * x' + φ)
```

Where:
- `(x, y)` are the coordinates of a point in the image.
- `f` is the frequency of the sinusoidal factor.
- `θ` is the orientation of the normal to the parallel stripes of the Gabor function.
- `σ` is the standard deviation of the Gaussian envelope.
- `γ` is the spatial aspect ratio.
- `φ` is the phase offset.

## Usage

The Gabor filter is commonly used in image processing libraries such as OpenCV and scikit-image. It is implemented using a convolution operation between the Gabor kernel and the input image.

For more information, you can refer to the [Wikipedia page on Gabor filters](https://en.wikipedia.org/wiki/Gabor_filter).



````python
import cv2
import numpy as np
import matplotlib.pyplot as plt

path = r"D:\Research\Git\CV-ImageProcessing\image\inputCT.png"
input_image  = cv2.imread(path, 0)


ksize = 5
sigma = 2
theta = 0
lambd = 1*np.pi/4
gamma = 0.25
phi = 6

kernel = cv2.getGaborKernel((ksize, ksize), sigma, theta, lambd, gamma, phi, ktype=cv2.CV_32F)

g_img = cv2.filter2D(input_image, cv2.CV_8UC3, kernel)

plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
plt.imshow(cv2.cvtColor(input_image, cv2.COLOR_BGR2RGB))
plt.title("Original Image")
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(cv2.cvtColor(g_img, cv2.COLOR_BGR2RGB))
plt.title("Gabor Filter Output")
plt.axis('off')

plt.show()
````
