# Histogram Equalization

Histogram equalization is a method in image processing of contrast adjustment using the image's histogram. This technique improves the contrast in an image by spreading out the most frequent intensity values, enhancing the global contrast of the image, especially when the usable data of the image is represented by close contrast values.

## Usage

Histogram equalization is widely used in image processing to enhance the contrast of images. It can be particularly effective for images that are dark or low contrast.
```python
import cv2
import matplotlib.pyplot as plt

path = r"D:\Research\Git\Image-Processing\image\inputCT.png"
segmented_image = cv2.imread(path, 0)
equ_image = cv2.equalizeHist(segmented_image)

plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.imshow(segmented_image, cmap='gray')
plt.title('Original Image')
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(equ_image, cmap='gray')
plt.title('Equalized Image')
plt.axis('off')

plt.show()
```
